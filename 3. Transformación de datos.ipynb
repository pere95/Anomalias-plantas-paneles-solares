# 2.CREACION DE VARIABLES

# 2.1 CreaciÃ³n eficiencia

```python
def eficiencia_inverter(AC,DC):
    t=(AC/DC)*100
    return(t.fillna(0))
```

```python
df['eficiencia']=eficiencia_inverter(df.kw_ac,df.kw_dc)
```

```python
df.eficiencia.isna().sum()
```

# 2.2 CreaciÃ³n componentes fecha

```python
def comp_fecha(dataf):
    dia=df.index.day
    mes=df.index.month
    hora=df.index.hour
    minuto=df.index.minute

    return(pd.DataFrame({'mes':mes,'dia':dia,'hora':hora,'minuto':minuto}))
```

```python
temporal=comp_fecha(df)
temporal
```

```python
df=pd.concat([df.reset_index(),temporal],axis=1)
df
```

```python
df.set_index('fecha',inplace=True)
df.isna().sum()
```

```python
orden = ['planta','mes','dia','hora','minuto','sensor_id','irradiacion','t_ambiente','t_modulo','inverter_id','kw_dc','kw_ac','eficiencia','kw_dia','kw_total']
```

```python
df=df[orden]
df
```

# 2.3 CreaciÃ³n dataframe diario (resampling)

```python
df.select_dtypes('number').columns
```

```python
df.columns[~df.columns.isin(df.select_dtypes('number').columns)]
```

```python
#no se incluye sensor_id porque cada planta tiene un unico sensor(valor constante)

df_dia = df.groupby(['planta', 'inverter_id']).resample('D').agg({
    'irradiacion': ['min', 'mean', 'max'],
    't_ambiente': ['min', 'mean', 'max'],
    't_modulo': ['min', 'mean', 'max'],
    'kw_dc': ['min', 'mean', 'max', 'sum'],
    'kw_ac': ['min', 'mean', 'max', 'sum'],
    'eficiencia': ['min', 'mean', 'max'],
    'kw_dia': 'max',
    'kw_total': 'max'
})

df_dia
```

```python
#Para eliminar el formato multiindice:
tuplas=df_dia.columns.to_flat_index()
tuplas
```

```python
df_dia.columns=['_'.join(par) for par in tuplas]
df_dia
```

```python
df_dia.reset_index().set_index('fecha')
```
